```python
import pandas as pd
import numpy as np

spam_words = pd.read_csv("./spammer_keyword_list.csv", encoding='cp949', index_col=0, squeeze=True).to_dict()
print(spam_words)
```

```python
ScreenName = []
Text = []
Degree = []
Ratio = []
Score = []
SpamScore = []

import re
score2 = 0
try:
        account = input()
        #print(account)
        timeline = twitter_api.GetUserTimeline(screen_name=account, count=200, include_rts=True, exclude_replies=False)
        #timeline2 = twitter_api.GetUserTimeline(screen_name=account, count=200, include_rts=False, exclude_replies=False)
        cnt = 0

        tweet = []
        retweet = []
        ScrenName2 = []
        Degree2 = []
        for t in timeline:
            ScreenName.append(account)
            cnt += 1
            #print(k, account, cnt, '--*'*30)
            txt = t.text.split()
            txt.pop()
            txt = " ".join(txt)
            new_text = re.sub("[^가-힣\n ]","",txt).split(" ")

            #print(new_text)

            score = 0
            for x in new_text:
                if x in spam_words:
                    score += spam_words[x]
            score = round(score,4)
            score2 += score
            Score.append(score)

            spamscores = round(score2 / len(Score),4)

            if len(new_text) == 0:
                Text.append("none")
            else:
                Text.append(new_text)

        Score.clear()
        SpamScore.append(spamscores)

        print("점수 : {}".format(spamscores))

except twitter.TwitterError as twerror:
        SpamScore.append(0)
```
